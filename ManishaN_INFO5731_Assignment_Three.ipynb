```python
# Write your code here
# 2.1 To build the documents-terms weights (tf*idf) matrix bold text
from sklearn.feature_extraction.text import TfidfVectorizer
dataFrame = open('movieReviews.csv',"r")
# initialising vector
vector = TfidfVectorizer()
response = vector.fit_transform(dataFrame)
# data frame with frequency
updatedDataFrame = pd.DataFrame(response.toarray(), columns=vector.get_feature_names())
updatedDataFrame


```

```python
# measuring similarity between two sentences using cosine similarity. 
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
# defining cos similarity 
def cosSimilarity(X,Y):
  m
# tokenize lists
  listX = word_tokenize(X) 
  listY = word_tokenize(Y)
# finding stopwords
  stopWords = stopwords.words('english') 
  list1 =[];list2 =[] 
  setX = {w for w in listX if not w in stopWords} 
  setY = {w for w in listY if not w in stopWords} 
# combining both the sets
  vector = setX.union(setY) 
  for w in vector: 
    if w in setX: list1.append(1)
    else: list1.append(0) 
    if w in setY: list2.append(1) 
    else: list2.append(0) 
#initialising count to 0
  count = 0

# cosine formula 
  for i in range(len(vector)): 
      count+= list1[i]*list2[i] 
# calculating cosine value
  cosine = count / float((sum(list1)*sum(list2))**0.5)
  return cosine
```
