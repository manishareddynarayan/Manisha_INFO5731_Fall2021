```python
import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from selenium import webdriver
import time
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

# setup driver and provide url to it
options = webdriver.ChromeOptions()
url = "https://www.imdb.com/title/tt9376612/reviews?ref_=tt_sa_3"
driver.get(url)
time.sleep(10)
page = 0
# assigning empty arrays to collect data
names = []
stars = []
reviewTitles = []
reviewTexts = []
postedDates = []
while True:
# get data from the webpage using beautifulsoup
    time.sleep(10)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    movies_rew_data = soup.find_all("div", {"class": "review-container"})
# considering till page 40 to retrive 1000 records with the attributes needed
    if(page==30) :
      for container in movies_rew_data:
          names.append(container.find('span',{"class":"display-name-link"}).get_text().strip().replace('\n',''))
          stars.append(container.span.text.replace('\n','').replace('\r',''))
          reviewTitles.append(container.find("a", {"class": "title"}).get_text().strip().replace('\n',''))
          postedDates.append(container.find("span", {"review-date"}).get_text().strip().replace('\n',''))
          reviewTexts.append(container.find("div", {"class": "content"}).get_text().strip().replace('\n',''))
# this is to load more data on clicking the load more button on bottom of the page
    if len(driver.find_elements_by_css_selector('.load-more-data')) > 0:
          driver.find_element_by_css_selector('.load-more-data').click()
          page+=1
          if int(page)>30:
           break
    else:
          break

df = pd.DataFrame(list(zip(reviewTitles, reviewTexts)), columns =['Title', 'Review'])
print("Length of data frame is {0}".format(len(df)))
df
```

```python
# Write your code here
import pandas as pd
# reading the provided txt file
info = pd.read_csv('surveyfile.txt',sep='\t\n')
info.columns=['colmn']
data=info
data.head()
# Printing top 5 results for reference
```

```python
# To find number of Words
info['word_count'] = info['colmn'].apply(lambda x: len(str(x).split(" ")))
info[['colmn','word_count']].head()
```

```python
# To find Number of characters 
info['character_count'] = info['colmn'].str.len() ## this also includes spaces
info[['colmn','character_count']].head()
```

```python
# Calculating Average Word Length
def avg_word(sentence):
# Splitting the words in sentence
  words = sentence.split()
  return (sum(len(word) for word in words)/len(words))
info['avg_word_count'] = info['colmn'].apply(lambda x: avg_word(x))
info[['colmn','avg_word_count']].head()
```

```python
# importing ntlk and downloading stopwords to find number of stopwords
import nltk
nltk.download('stopwords')
```

```python
# Calculating Number of stopwords

from nltk.corpus import stopwords
stop = stopwords.words('english')

info['stop_words'] = info['colmn'].apply(lambda x: len([x for x in x.split() if x in stop]))
info[['colmn','stop_words']].head()
```

```python
# Calculating Number of special characters

info['spl_char'] = info['colmn'].apply(lambda x: len([x for x in x.split() if x.startswith('#')]))
info[['colmn','spl_char']].head()
```

```python
# Solvinf for number of numerics
info['num_numerics'] = info['colmn'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
info[['colmn','num_numerics']].head()
```

```python
# Solving for words with all Uppercase letters in it
info['no._uppercase'] = info['colmn'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
info[['colmn','no._uppercase']].head()
```

```python
# Changing all letters to Lower case

info['colmn'] = info['colmn'].apply(lambda x: " ".join(x.lower() for x in x.split()))
info['colmn'].head()
```

```python
# To remove all the Punctuations
info['colmn'] = info['colmn'].str.replace('[^\w\s]','')
info['colmn'].head()
```

```python
# Removing of Stop Words by importing stopwords
from nltk.corpus import stopwords
stop = stopwords.words('english')
info['colmn'] = info['colmn'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
info['colmn'].head()
```

```python
# To find commonly occurring words
freq = pd.Series(' '.join(info['colmn']).split()).value_counts()[:10]
freq
```

```python
# To remove commonly occurring words
freq = list(freq.index)
info['colmn'] = info['colmn'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
info['colmn']
```

```python
# To find Rare words
freq = pd.Series(' '.join(info['colmn']).split()).value_counts()[-10:]
freq
```

```python
# To remove Rare words
freq = list(freq.index)
info['colmn'] = info['colmn'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
info['colmn']
```

```python
# To make spelling corrections
from textblob import TextBlob
info['colmn'][:5].apply(lambda x: str(TextBlob(x).correct()))
```

```python
# Dividing the text into a sequence of words(Tokenization) by importing nltk and downloading punkt
import nltk
nltk.download('punkt')
TextBlob(info['colmn'][4]).words
```

```python
# removing of suffices(Stemming)
from nltk.stem import PorterStemmer
st = PorterStemmer()
info['colmn'][:5].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
```

```python
# converting the word into its root word(Lemmatization)
from textblob import Word
import nltk
nltk.download('wordnet')
info['colmn'] = info['colmn'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
info['colmn'].head()
```

```python
# Write your code here
from nltk.tokenize import word_tokenize
import nltk
nltk.download('averaged_perceptron_tagger')

pos = []
for sentence in df1['colmn']:
  text = word_tokenize(str(sentence))
  pos.append(nltk.pos_tag(text))

nouns=0
verbs=0
adverbs=0
adjectives=0
for i in pos:
  for j in i:
    if 'NN' in j:
      nouns+=1
    if 'VB' in j:
      verbs+=1
    if 'ADJ' in j:
      adjectives+=1
    if 'ADV' in j:
      adverbs+=1
print("nouns:",nouns)  
print("verbs:",verbs) 
print("adverbs:",adverbs) 
print("adjectives:",adjectives)
pos
```

```python
# You explaination here
In constituency parsing the given sentence is broken into subphrases that belong to specific category in grammer.
It is based on the formalism of context-free grammars. It always contains words of the sentence smiliar to its terminal nodes.
All remaining non-terminal nodes represent constituents of the sentence and mainly are one of  noun phrase, verb phrase or prepositional phrase.
Dependency parsing tree graph is the set of vertices that contain the words in given sentence, and each edge connects two words.
It does not make use of sub-phrases, instead dependencies in the sentence are expressed.

```

```python
# Write your code here

def removeLeadZeros(ip):
    # splits given ip and converts the words to remove the leading zeros. Once done convert them back to string and join
    new_ip = ".".join([str(int(i)) for i in ip.split(".")])  
    return new_ip ;
      
# Use deifned function
ip ="260.08.094.109"
print("The ip address with all the leading zeros removed: " + str(removeLeadZeros(ip)))


```
