```python
# You code here (Please add comments in the code):

import pandas as pd
# reading the provided txt file
info = pd.read_csv('surveyfile.txt',sep='\t\n')
info.columns=['colmn']
data=info
data
```

```python
# 1.Term Freqency
import numpy as np
termfreq = (info['colmn'][0:10]).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
termfreq.columns = ['words','termfreq']
termfreq
# 2.Inverse Document Frequency
# enumerating the term frequency of words from above to solve inverse document frequency
for i,word in enumerate(termfreq['words']):
  termfreq.loc[i,'inversedocfreq'] = np.log(info.shape[0]/len(info[info['colmn'].str.contains(word)]))
termfreq
# Term Frequency – Inverse Document Frequency (TF-IDF)
termfreq['tfidf'] = termfreq['termfreq'] * termfreq['inversedocfreq']
termfreq
```

```python
# 3.N-grams 
import nltk
# Specifying to get unigram
one_gram = (pd.Series(nltk.ngrams(info['colmn'],1)).value_counts())[:10]
print("Printing unigram" + str(one_gram))
# Specifying to get bigram
two_gram = (pd.Series(nltk.ngrams(info['colmn'],2)).value_counts())[:10]
print("Printing bigram" + str(two_gram))
```

```python
# 4.Bag-of-words
# Importing count vectorizer to find the count
from sklearn.feature_extraction.text import CountVectorizer
coun = CountVectorizer(max_features=50, lowercase=True, ngram_range=(1,3), analyzer = "word")
ex = coun.fit_transform(info['colmn'])
# get_feature_names() function is used to extract words
coun.get_feature_names()
```

```python
# 5.Named entity recognition
from nltk.tokenize import word_tokenize
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
words = []
#taking values from info with column colmn
for sent in info['colmn']:
  text = word_tokenize(str(sent))
  words.append(nltk.pos_tag(text))
words
```
